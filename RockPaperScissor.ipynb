```python
import random
keepgoing='y'
tie=0
win=0
lose=0
while(keepgoing=='y'):
    #For user
    print("choose Rock,Paper,Scissor")
    a=input()
    #For computer
    c=['Rock','Paper','Scissor']
    b=random.choice(c)
    print(b)
    print("User choosen: "+a+"\nComputer choosen: "+b)
    if(a=="Rock" and b=="Scissor"):
        print("User is the winner.")
        win=win+1
    elif(a=="Scissor" and b=="Paper"):
        print("User is the winner.")
        win=win+1
    elif(a=="Paper" and b=="Rock"):
        print("User is the winner.")
        win=win+1
    elif(a=="Scissor" and b=="Rock" ):
        print("User is the lose.")
        lose=lose+1
    elif(a=="Paper" and b=="Scissor"):
        print("User lose.")
        lose=lose+1
    elif(a=="Rock" and b=="Paper"):
        print("User lose.")
        lose=lose+1
    else:
        print("It's Tie.")
        tie=tie+1
    print("Do You want to play another round??")
    keepgoing=input()

total=win+lose+tie
print("User's win=",win)
print("User's Lose=",lose)
print("User's Tie=",tie)
print("System's win=",total-(lose+tie))
print("System's Lose=",total-(win+tie))
print("System's Tie=",tie)
```
